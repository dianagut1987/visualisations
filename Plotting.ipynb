```
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2,'g',label='quadratic')  # etc.
plt.plot(x, x**3,'m', label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(40, 0.015,r'$\frac{ \beta }{ \alpha } - \frac{\partial^nf}{\partial x^n}$',fontsize=25)#r before the string - raw string
#for tex formulas: http://atomurl.net/math/
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.annotate('', (60, 0.01),
            xytext=(50, 0.01),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=16,
            horizontalalignment='right', verticalalignment='top')

plt.grid(True)
```

```
x=np.linspace(1,10**9,1000)
y=x**-2
plt.loglog(x, y)
plt.title('Linear')
plt.grid(True)

plt.axis([1, 10**9, 0, 1])
```

```
x=np.linspace(1,10**9,1000)
y=x**(-2)
plt.subplot(311)
plt.plot(x, y)
plt.title('Linear')
plt.grid(True)
plt.axis([0, 10**9, 0, 1])

plt.subplot(312)
plt.loglog(x, y)
plt.title('Loglog')
plt.grid(True)


plt.subplot(313)
plt.plot(x, y)
plt.title('semi-Log')
plt.xscale("log")
plt.grid(True)
```

```

import matplotlib.pyplot as plt
import numpy as np

ax = plt.figure().add_subplot(projection='3d')


X, Y = np.mgrid[0:6*np.pi:0.25, 0:4*np.pi:0.25]
Z = np.sqrt(np.abs(np.cos(X) + np.cos(Y)))

ax.plot_wireframe(X + 1e5, Y + 1e5, Z,azim=10)

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")
ax.set_zlim(0, 2)

plt.show()
```
